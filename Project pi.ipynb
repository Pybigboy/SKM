<b>с 2, 4, 8, 16, 32, 64 и 128 узлами, найти приближенное к  π</b>



```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import integrate
```

С помощью данного интеграла можно найти приближенное к $\pi$

$$ \pi = \int_0^1\frac{4}{1+x^2}dx $$

График этой функции, выглядит следующий образом:


```python
x = np.arange(0, 1, 0.001)
y = 4 / (1 + x**2)

plt.plot(x, y)
plt.grid()
```

# Используем правило *прямоугольников* с 2, 4, 8, 16, 32, 64 и 128 узлами и выведем погрешность

```python
uzli = np.array([2**i for i in range(1,8)])

a = 0
b = 1

for uzel in uzli:
    
    delta_x = (b - a) / uzel
    
    x = np.arange( (a + (delta_x / 2)) , b, delta_x)
    y = 4/(1+x**2)
    
    sum_storon = sum(y)
    
    rect_pr = sum_storon * delta_x
    razn = abs(np.pi - rect_pr) / np.pi
    
    rect_method = plt.plot(uzel, razn)
    plt.setp(rect_method, marker = 'o', color = 'tab:blue')
    
    print('Узлов:', uzel, '\tПогрешность:', razn)
    
rect_pr_pogresh = razn

plt.grid()
plt.yscale('log')
plt.xscale('log')

rect_pr
```

# Используем правило *трапеций*  с 2, 4, 8, 16, 32, 64 и 128 узлами и выведем погрешность

```python
uzli = np.array([2**i for i in range(1,8)])

a = 0
b = 1

for uzel in uzli:
    
    x = np.linspace(a, b, uzel + 1)
    y = 4 / ( 1 + x**2 )
    
    trap_pr = integrate.trapz(y, x)
    razn = (np.pi - trap_pr) / np.pi
    
    trap_method = plt.plot(uzel, razn)
    plt.setp(trap_method, marker = 'o', color = 'tab:red')
    
    print('Узлов:', uzel, '\tПогрешность:', razn)
    
trap_pr_porgesh = abs(razn)

plt.grid()
plt.yscale('log')
plt.xscale('log')

trap_pr
```

# Используем правило *Симпсона*  с 2, 4, 8, 16, 32, 64 и 128 узлами и выведем погрешность

```python
uzli = np.array([2**i for i in range(1,8)])

a = 0
b = 1

for uzel in uzli:
    
    delta_x = (b - a) / uzel
    
    x = np.linspace(a, b, uzel + 1)
    y = 4 / ( 1 + x**2 )
    
    simps_pr = integrate.simps(y) * delta_x
    razn = abs(np.pi - simps_pr) / np.pi
    
    simpson_method = plt.plot(uzel, razn)
    plt.setp(simpson_method, marker = 'o', color = 'tab:pink')
    
    print('Узлов:', uzel, '\tПогрешность:', razn)
    
simps_pr_pogresh = razn

plt.grid()
plt.yscale('log')
plt.xscale('log')

simps_pr
```

```python
print('Методы\t', '\t\t\t\t Погрешность')
print('Прямоугольников:', rect_pr , '\t',rect_pr_pogresh )
print('Трапеций:\t', trap_pr , '\t',trap_pr_porgesh )
print('Симпсона:\t', simps_pr , '\t',simps_pr_pogresh )
print('PI:\t\t', np.pi)
```

 * Правило Симпсона - погрешность уменьшается приблизительно в 16 раз
